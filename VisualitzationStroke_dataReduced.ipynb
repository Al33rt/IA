# Visualitzation of stroke dataset

### Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('data/clean_stroke.pkl')
pd.set_option("display.precision", 2)
```

```python
df.head()
```

### Visualtization

```python
sns.set_theme()
```

```python
sns.displot(data=df, x="gender", hue="stroke", multiple="dodge")
```

```python
sns.pairplot(df, hue='stroke')
```

```python
sns.catplot(data=df, kind = "violin", x="stroke", y="age")
```
