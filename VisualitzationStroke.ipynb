# Visualitzation of stroke dataset

### Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('data/clean_stroke.pkl')
pd.set_option("display.precision", 2)
```

```python
df.head()
```

### Visualtization

```python
sns.set_theme()
```

```python
sns.pairplot(df, hue='stroke', diag_kind="hist")
```

```python
sns.displot(data=df, x="gender", hue="stroke", multiple="dodge")
```

Com ja havíem vist tenim moltes més dades de casos negatius que positius, addicionalment podem veure que tenim més dades de dones que homes, però res fora del normal.

```python
fig, axs = plt.subplots(ncols=2)
sns.violinplot(data=df, x="stroke", y="age",ax=axs[0])
sns.boxplot(data=df, x="stroke", y="age", ax=axs[1])
```

Aquí podem veure que en els casos negatius tenim unes dades bastant homogènies, des de nadons a vells. I en els casos positius la majoria de dades són de gent adulta, principalment entre els 60 i 80 anys, amb dos casos puntals de nens. 

```python
sns.displot(data=df, x="hypertension", hue="stroke", multiple="dodge")
```

Amb aquest gràfic es veu que la majoria de les dades han estat obtingudes de gent que no ha patit hipertensió.

```python
sns.displot(data=df, x="heart_disease", hue="stroke", multiple="dodge")
```

I aquí ens torna a passar com en el cas anterior on quasi totes les dades són de gent que no ha patit malalties de cor anteriorment.

```python
sns.displot(data=df, x="ever_married", hue="stroke", multiple="dodge")
```

```python
sns.displot(data=df, x="work_type", hue="stroke", multiple="dodge", height=5, aspect=15/8)
```

```python
sns.displot(data=df, x="Residence_type", hue="stroke", multiple="dodge")
```
